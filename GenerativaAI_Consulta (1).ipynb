```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.embeddings import HuggingFaceEmbeddings  # Para embeddings open source
from langchain_community.vectorstores import Chroma
from langchain.chains import RetrievalQA  # Para QA sobre documentos
from langchain_community.llms import Ollama  # Exemplo de LLM local open source
```

```python
# 1. Carregar o arquivo PDF (mantemos o PyPDFLoader, que é open source)
loader = PyPDFLoader("Orgulho e preconceito.pdf")
pages = loader.load_and_split()
```

```python
# 2. Escolher um modelo de embeddings open source (HuggingFace Transformers)
# Você pode escolher diferentes modelos pré-treinados. "all-MiniLM-L6-v2" é um popular.
embeddings_model_name = "all-MiniLM-L6-v2"
embeddings = HuggingFaceEmbeddings(model_name=embeddings_model_name)
```

```python
# 3. Criar o banco de dados vetorial com Chroma (open source)
vectordb = Chroma.from_documents(pages, embedding=embeddings,
                                 persist_directory="./chroma_db_opensource")
vectordb.persist()
```

```python
# 4. Escolher um Large Language Model (LLM) open source
# Aqui, usamos Ollama como exemplo. Ollama permite executar LLMs localmente.
# Você precisará ter o Ollama instalado e um modelo compatível rodando.
# Exemplo de modelo que você pode baixar com Ollama: llama2
llm = Ollama(model="llama2")
```

```python
# Outras opções de LLMs open source (requerem configuração diferente):
# - HuggingFace Hub Inference API (alguns modelos são gratuitos ou de baixo custo)
# - Modelos executados localmente via Transformers (mais complexo de configurar)

# 5. Criar a cadeia de Question Answering (RetrievalQA é uma alternativa open source a ChatVectorDBChain)
qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",  # Outras opções: map_reduce, refine, map_rerank
    retriever=vectordb.as_retriever(search_kwargs={"k": 3}),  # Ajuste o número de documentos a serem buscados
    return_source_documents=True
)
```

```python
# 6. Fazer a pergunta
query = "What are Elizabeth Bennet's initial impressions of Mr. Darcy?"
result = qa({"query": query})
```

```python
# 7. Imprimir a resposta
print("Answer:")
print(result["result"])
```
